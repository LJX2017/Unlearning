```python
import re

# Example string

for i in range(len(prompts)):
    text = prompts[i]['eval']
    match = re.search(r"MODEL_FAMILIARITY\s*:\s*(\d+)/3", text)
    if match:
        y = match.group(1)  # Extract the value of Y
        prompts[i]['score'] = y
    else:
        print("No match found")
        print(i)
        print(text)
```
